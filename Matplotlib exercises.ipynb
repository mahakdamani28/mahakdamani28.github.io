***1.line chart***
_Visualize and compare monthly sales trends of two products using line chart.

```python
import matplotlib.pyplot as plt
```

```python
# Sample Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_A = [30, 35, 40, 42, 45, 50]
sales_B = [25, 28, 35, 37, 40, 48]
```

```python
# Create line chart for product A
plt.plot(months, sales_A,label='Product A', color='blue', marker='o', linestyle='-')

# Create line chart for product B
plt.plot(months, sales_B,label='Product B', color='green', marker='s', linestyle='--')

# Add titles and labels
plt.title('Monthly Sales Comparison')
plt.xlabel('Months')
plt.ylabel('Sales (rs.1000)')

# Add legend and grid
plt.legend()
plt.grid(True)

# Display the chart
plt.show()
```

***2. Bar Chart*** _Visualize student enrollment in different courses using a bar chart

```python
import matplotlib.pyplot as plt
```

```python
# Sample Data: Courses and student enrollments
courses = ['Python', 'Data Science', 'Web Dev', 'AI/ML', 'Cyber Security']
students = [120, 95, 75, 110, 60]
```

```python
# Create the bar chart
plt.bar(courses, students, color='green', edgecolor='black')

# Add titles and labels
plt.title('Student Enrollment in different Courses')
plt.xlabel('Courses')
plt.ylabel('Number of Students')

# Add grid and show the chart
plt.grid(axis='y')
plt.show()
```

***3. Histogram:***
Visualize how people are distributed across different age groups


```python
import matplotlib.pyplot as plt
```

```python
# Sample data: ages of 25 people
ages = [18, 21, 22, 25, 27, 29, 30, 32, 34, 36,
        37, 39, 40, 41, 43, 45, 47, 48, 50, 52,
        54, 55, 57, 59, 60]
```

```python
# Create histogram
plt.hist(ages, bins=7, color='orange', edgecolor='black')

# Add titles and labels
plt.title('Age Distribution of People')
plt.xlabel('Age Group')
plt.ylabel('Number of People')

# Show grid and plot
plt.grid(axis='y')
plt.show()
```

***4.Scatter plot:*** Visualize how temperature(in celcius) affects ice cream sales (in rs.1000s)

```python
import matplotlib.pyplot as plt
```

```python
# Sample data: temperature (celcius) vs. ice cream sales (in rs.1000s)
temperature = [20, 22, 24, 26, 28, 30, 32, 34, 36, 38]
sales = [ 10, 12, 15, 18, 22, 26, 30, 35, 40, 45 ]
```

```python
# Create a scatter plot
plt.scatter(temperature,sales, color='green', marker='^'  )

# Add titles and labels
plt.title('Temperature vs. Ice Cream Sales')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Ice Cream Sales (rs.1000s)')

# Show grid and plot
plt.grid(True)
plt.show()
```
