```python
# Step 1: Import Library
import pandas as pd
```

```python
# Step 2: Read Dataset
df = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/refs/heads/main/Ice%20Cream.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Step 3: Define y and x
df.columns
```

```python
y = df['Revenue']
X = df[['Temperature']]
```

```python
df.shape , X.shape , y.shape
```

```python
# step 4: Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=2529)
```

```python
# step 5: Select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# step 6: Train model
model.fit(X_train, y_train)
```

```python
model.intercept_ , model.coef_
```

```python
# step 7: Prediction
y_pred = model.predict(X_test)
```

```python
# step 8: Evaluate
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
# Step 1: Import Library
import pandas as pd
```

```python
# STep 2: Read dataset
df = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/refs/heads/main/Salary%20Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# step 3: define y and x
df.columns
```

```python
y = df['Salary']
X = df[['Experience Years']]
```

```python
df.shape , X.shape , y.shape
```

```python
# Step 4: Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.75, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5: Select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Step 6: Train model
model.fit(X_train, y_train)
```

```python
model.intercept_ , model.coef_
```

```python
#Step 7: Prediction
y_pred = model.predict(X_test)
```

```python
#Step 8: Evaluate
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred)
plt.plot(y_test, y_test, color='red')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```
