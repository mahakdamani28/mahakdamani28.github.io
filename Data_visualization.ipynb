1. **Line Chart**
objective: Visualize monthly sales (in rs.1000s) over a year using line chart

```python
import matplotlib.pyplot as plt
```

```python
# Sample Data: Months vs Sales
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales =[25, 28, 30, 35, 40, 38, 45, 50, 48, 42, 38, 33]
```

```python
# Create the line chart
plt.plot(months, sales, color='blue', marker='o', linestyle='-', linewidth=2)
```

```python
# Add title and axis labels
plt.title('Monthly Sales Trend (in Rs.1000s)')
plt.xlabel('Months')
plt.ylabel('Sales (in Rs.1000s)')

# Add grid and display the plot
plt.grid(True)
plt.show()
```

**2. Bar Chart:**
Visualize the sales of different products using a bar chart

```python
import matplotlib.pyplot as plt
```

```python
# Sample data: Product names and their sales
products = ['Laptop', 'Tablet', 'Smartphone', 'Headphones', 'Smartwatch']
sales = [150, 100, 300, 120, 80]
```

```python
# Create a bar chart
plt.bar(products, sales, color='skyblue', edgecolor='black')

# Add title  and axis labels
plt.title('Sales of Products (in units)')
plt.xlabel('Products')
plt.ylabel('Units Sold')

# Add grid and display the plot
plt.grid(axis='y')
plt.show()
```

**Histogram:**
Visualize the distribution of students' marks uding histogram


```python
import matplotlib.pyplot as plt
```

```python
# Sample data: marks of 30 students
marks =[ 45, 56, 67, 49, 78, 82, 91, 55, 62, 74, 68, 85, 89, 73, 50, 66, 77, 88, 93, 59, 61, 70, 72, 64, 79, 81, 60, 69, 90, 76 ]
```

```python
# Create the histogram
plt.hist(marks, bins=8, color='skyblue', edgecolor='black')

# Add title and axis labels
plt.title('Distribution of Student Marks')
plt.xlabel('Marks Range')
plt.ylabel('Number of Students')

# Add grid and display the plot
plt.grid(axis='y')
plt.show()
```
